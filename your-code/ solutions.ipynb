```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"]="/home/martin/Ironhack - Labs/Ironhack Data Analytics.json"
from google.cloud import bigquery
client = bigquery.Client()
```

```python
challenge1="""
SELECT
    authors.au_id,
    authors.au_fname,
    authors.au_lname,
    book_titles.title,
    publishers.pub_name
    
FROM
    `ironhack-data-analytics.publications.authors` authors
    
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
ON
    authors.au_id = title_author.au_id
    
INNER JOIN
    `ironhack-data-analytics.publications.titles` book_titles
ON
    title_author.title_id = book_titles.title_id
    
INNER JOIN
    `ironhack-data-analytics.publications.publishers` publishers
ON
    book_titles.pub_id = publishers.pub_id

GROUP BY
    1,
    2,
    3,
    4,
    5
    

"""
```

```python
query_job=client.query(query=challenge1)
df=query_job.to_dataframe()
df
```

```python
challenge2="""
SELECT
    authors.au_id,
    authors.au_fname,
    authors.au_lname,
    publishers.pub_name,
    COUNT(book_titles.pub_id) as many_titles
    
FROM
    `ironhack-data-analytics.publications.authors` authors
    
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
ON
    authors.au_id = title_author.au_id
    
INNER JOIN
    `ironhack-data-analytics.publications.titles` book_titles
ON
    title_author.title_id = book_titles.title_id
    
INNER JOIN
    `ironhack-data-analytics.publications.publishers` publishers
ON
    book_titles.pub_id = publishers.pub_id

GROUP BY
    authors.au_id,
    authors.au_fname,
    authors.au_lname,
    publishers.pub_name
    

"""
```

```python
query_job=client.query(query=challenge2)
df=query_job.to_dataframe()
df
```

```python
challenge3="""
SELECT
    authors.au_id,
    authors.au_fname,
    authors.au_lname,
    SUM(sales.qty) as Total
    
FROM
    `ironhack-data-analytics.publications.authors` authors
    
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
ON
    authors.au_id = title_author.au_id
    
INNER JOIN
    `ironhack-data-analytics.publications.sales` sales
ON
    title_author.title_id = sales.title_id

GROUP BY
    authors.au_id,
    authors.au_fname,
    authors.au_lname

ORDER BY
    Total DESC
    
LIMIT
    3
"""
```

```python
query_job=client.query(query=challenge3)
df=query_job.to_dataframe()
df
```

```python
challenge4="""
SELECT
    authors.au_id,
    authors.au_fname,
    authors.au_lname,
    COALESCE(SUM(sales.qty), 0) as Total
    
FROM
    `ironhack-data-analytics.publications.authors` authors
    
LEFT JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
ON
    authors.au_id = title_author.au_id
    
LEFT JOIN
    `ironhack-data-analytics.publications.sales` sales
ON
    title_author.title_id = sales.title_id

GROUP BY
    authors.au_id,
    authors.au_fname,
    authors.au_lname

ORDER BY
    Total DESC
    
LIMIT
    23
"""
```

```python
query_job=client.query(query=challenge4)
df=query_job.to_dataframe()
df
```

```python
bonus="""
SELECT
    authors.au_id,
    authors.au_fname,
    authors.au_lname,
    (title.advance * title_author.royaltyper / 100 + title.price * sales.qty * title_author.royaltyper / 100 * title.royalty / 100) as Profit
    
FROM
    `ironhack-data-analytics.publications.authors` authors
    
INNER JOIN
    `ironhack-data-analytics.publications.titleauthor` title_author
ON
    authors.au_id = title_author.au_id
    
INNER JOIN
    `ironhack-data-analytics.publications.sales` sales
ON
    title_author.title_id = sales.title_id
    
INNER JOIN
    `ironhack-data-analytics.publications.titles` title
ON
    sales.title_id = title.title_id

ORDER BY
    Profit DESC
    
LIMIT
    3
"""
```

```python
query_job=client.query(query=bonus)
df=query_job.to_dataframe()
df
```
